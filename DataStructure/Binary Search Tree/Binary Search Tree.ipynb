```python

# BST
class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
def InsertNode(root,node):
    
    if (root.data is None):
        root=node
    else:
        if (root.data<node.data):
            
            if (root.right is None):
                root.right=node
            else:
                InsertNode(root.right,node)
        else:
            if (root.left is None):
                root.left=node
            else:
                InsertNode(root.left,node)
                

            
            
def PrintInorderTree(root):
    
    if root is None:
        return
    PrintInorderTree(root.left)
    print(root.data)
    PrintInorderTree(root.right)
        
if __name__=="__main__":
    s=Node(50)
    InsertNode(s,Node(60))
    InsertNode(s,Node(10))
    InsertNode(s,Node(40))
    InsertNode(s,Node(7))
    InsertNode(s,Node(5))
    InsertNode(s,Node(80))
    PrintInorderTree(s)
    
```

```python
# Searching Node IN BST

class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
def InsertNode(root,node):
    
    if (root.data is None):
        root=node
    else:
        if (root.data<node.data):
            
            if (root.right is None):
                root.right=node
            else:
                InsertNode(root.right,node)
        else:
            if (root.left is None):
                root.left=node
            else:
                InsertNode(root.left,node)
                
def SearchNode(root,skey):
    if root is None:
        return False
    else:
        if root.data==skey:
            return True
        if root.data<skey:
            return SearchNode(root.right,skey)
        return SearchNode(root.left,skey)
            
            
def PrintInorderTree(root):
    
    if root is None:
        return
    PrintInorderTree(root.left)
    print(root.data)
    PrintInorderTree(root.right)
        
if __name__=="__main__":
    s=Node(50)
    InsertNode(s,Node(60))
    InsertNode(s,Node(10))
    InsertNode(s,Node(40))
    InsertNode(s,Node(7))
    InsertNode(s,Node(5))
    InsertNode(s,Node(80))
    PrintInorderTree(s)
    if SearchNode(s,1):
        print('Found')
    else:
        print('Not Found')
```

```python
# Deletion Of Node In BST

class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
def InsertNode(root,node):
    
    if (root.data is None):
        root=node
    else:
        if (root.data<node.data):
            
            if (root.right is None):
                root.right=node
            else:
                InsertNode(root.right,node)
        else:
            if (root.left is None):
                root.left=node
            else:
                InsertNode(root.left,node)
                
def DeletionNode(root,dnode):
    
    if root is None:
        return root
    if root.data<dnode:
        root.right=DeletionNode(root.right,dnode)
    elif root.data>dnode:
        root.left=DeletionNode(root.left,dnode)
    else:
        if root.right is None:
            temp=root.left
            root=None
            return temp
        elif root.left is None:
            temp=root.right
            root=None
            return temp
        temp=Findminimum(root.right)
        root.data=temp.data
        root.right=DeletionNode(root.right,temp.data)
        
    return root

def Findminimum(node):
    current=node
    while(current.left is not None):
        current=current.left
    return current

            
           
            
def PrintInorderTree(root):
    
    if root is None:
        return
    PrintInorderTree(root.left)
    print(root.data,end=" ")
    PrintInorderTree(root.right)
        
if __name__=="__main__":
    s=Node(50)
    InsertNode(s,Node(60))
    InsertNode(s,Node(30))
    InsertNode(s,Node(40))
    InsertNode(s,Node(70))
    InsertNode(s,Node(20))
    InsertNode(s,Node(80))
    PrintInorderTree(s)
    DeletionNode(s,50)
    print()
    print("After Deletion:")
    PrintInorderTree(s)    
```
