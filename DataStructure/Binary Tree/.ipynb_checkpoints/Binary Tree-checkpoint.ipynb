```python
class Node():
    
    def __init__(self,data):
        
        self.right=None
        self.left=None
        self.data=data
        
        
def Inoreder(temp):
    if (not temp):
        return
    Inoreder(temp.left)
    print(temp.data)
    Inoreder(temp.right)
        
root=Node(1)

root.left=Node(2)
root.right=Node(3)
root.right.right=Node(4)
root.right.left=Node(5)
root.left.right=Node(6)
root.left.left=Node(7)

Inoreder(root)
```

```python
class BTreeNode():
    
    def __init__(self,data):
        
        self.left=None
        self.right=None
        self.data=data
        
class BTree():
    
    def __init__(self):
        
        self.root=None
        
        
    def AddNode(self,root,data):
        
        if root==None:
            self.root=BTreeNode(data)
            
        else:
            if(data<root.data):
                
                if(root.left==None):
                    root.left=BTreeNode(data)
                else:
                    self.AddNode(root.left,data)
            
            else:
                if (root.right==None):
                    root.right=BTreeNode(data)
                else:
                    self.AddNode(root.right,data)
                    
    def PrintInorder(self,root):
        if (root==None):
            return
        self.PrintInorder(root.left)
        print(root.data)
        self.PrintInorder(root.right)
        

bt=BTree()
bt.AddNode(bt.root,10)
bt.AddNode(bt.root,11)
bt.AddNode(bt.root,7)
bt.AddNode(bt.root,9)
bt.AddNode(bt.root,15)
bt.AddNode(bt.root,8)


bt.PrintInorder(bt.root)

                
```

```python
class BinaryTreeNode():
    
    def __init__(self,data):
        
        self.left=None
        self.right=None
        self.data=data
        
        
def InsertNode(root,node):
    if root==None:
        root=node
    else:
        if(node.data<root.data):
            if(root.left==None):
                root.left=node
            else:
                InsertNode(root.left,node)
        else:
            if(root.right==None):
                root.right=node
            else:
                InsertNode(root.right,node)
                
def FindNode(root,snode):
    
    if (root.data<snode.data):
        if root.left==None:
            return snode+'not found'
        else:
            return FindNode(root.left,snode)
    elif root.data>snode.data:
        if root.right==None:
            return snode+"not found"
        else:
            return FindNode(root.right,snode)
    else:
        print(root.data+'is Found')
            
        
    
                    
def Printinorder(root):
    if root==None:
        return
    Printinorder(root.left)
    print(root.data)
    Printinorder(root.right)
        
        
r=BinaryTreeNode(10) 
InsertNode(r,BinaryTreeNode(11))
InsertNode(r,BinaryTreeNode(7))
InsertNode(r,BinaryTreeNode(9))
InsertNode(r,BinaryTreeNode(15))
InsertNode(r,BinaryTreeNode(8))
InsertNode(r,BinaryTreeNode(17))
Printinorder(r)
FindNode(r,16)
        
```
