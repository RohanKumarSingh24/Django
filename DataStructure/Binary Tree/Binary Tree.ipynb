```python
class BTreeNode():
    
    def __init__(self,data):
        
        self.left=None
        self.right=None
        self.data=data
        
class BTree():
    
    def __init__(self):
        
        self.root=None
        
        
    def AddNode(self,root,data):
        
        if root==None:
            self.root=BTreeNode(data)
            
        else:
            if(data<root.data):
                
                if(root.left==None):
                    root.left=BTreeNode(data)
                else:
                    self.AddNode(root.left,data)
            
            else:
                if (root.right==None):
                    root.right=BTreeNode(data)
                else:
                    self.AddNode(root.right,data)
                    
    def PrintInorder(self,root):
        if (root==None):
            return
        self.PrintInorder(root.left)
        print(root.data)
        self.PrintInorder(root.right)
        

bt=BTree()
bt.AddNode(bt.root,10)
bt.AddNode(bt.root,11)
bt.AddNode(bt.root,7)
bt.AddNode(bt.root,9)
bt.AddNode(bt.root,15)
bt.AddNode(bt.root,8)


bt.PrintInorder(bt.root)

                
```

```python
class BinaryTreeNode():
    
    def __init__(self,data):
        
        self.left=None
        self.right=None
        self.data=data
        
        
def InsertNode(root,node):
    if root==None:
        root=node
    else:
        if(node.data<root.data):
            if(root.left==None):
                root.left=node
            else:
                InsertNode(root.left,node)
        else:
            if(root.right==None):
                root.right=node
            else:
                InsertNode(root.right,node)
                

            
        
def search(root,key): 
      
    
    if root is None or root.data == key: 
        return root 
  
     
    if root.data < key: 
        return search(root.right,key) 
    
     
    return search(root.left,key) 

        
        
                    
                    
                   
        
                    
def Printinorder(root):
    if root==None:
        return
    Printinorder(root.left)
    print(root.data)
    Printinorder(root.right)
        
        
r=BinaryTreeNode(10) 
InsertNode(r,BinaryTreeNode(11))
InsertNode(r,BinaryTreeNode(7))
InsertNode(r,BinaryTreeNode(9))
InsertNode(r,BinaryTreeNode(15))
InsertNode(r,BinaryTreeNode(8))
InsertNode(r,BinaryTreeNode(17))
Printinorder(r)
if search(r,21):
    print('found')
else:
    print('not found')

        
```
