```python
# Merge
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
   
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
print(df1, '\n\n', df2)
res=pd.merge(df1,df2,on='key')
res
```

```python
# join with multiple key
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
print(df1,'\n\n', df2)
r=pd.merge(df1,df2,on=['key','key1'])
r
```

```python
# using of how()
```

```python
#left join
```

```python
print(df1,'\n\n',df2)
h1=pd.merge(df1,df2,on=['key','key1'],how='left')
h1
```

```python
# right join
```

```python
print(df1,'\n\n',df2)
r=pd.merge(df1,df2,on=['key','key1'],how='right')
r
```

```python
# inner join
```

```python
print(df1,'\n\n',df2)
i=pd.merge(df1,df2,on=['key','key1'],how='inner')
i
```

```python
# outer join
```

```python
print(df1,'\n\n',df2)
o=pd.merge(df1,df2,on=['key','key1'],how='outer')
o
```
